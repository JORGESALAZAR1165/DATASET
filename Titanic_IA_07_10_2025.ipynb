PassengerId: ID único de cada pasajero.

Survived: Variable objetivo. 0 = No sobrevivió, 1 = Sí sobrevivió.

Pclass: Clase del boleto (1 = Primera, 2 = Segunda, 3 = Tercera).

Name: Nombre del pasajero.

Sex: Género del pasajero (male, female).

Age: Edad del pasajero. Puede tener datos faltantes.

SibSp: Número de hermanos/cónyuges a bordo.

Parch: Número de padres/hijos a bordo.

Ticket: Número de boleto.

Fare: Tarifa pagada por el pasaje.

Cabin: Número de cabina. Muchos valores faltantes.

Embarked: Puerto de embarque (C = Cherbourg, Q = Queenstown, S = Southampton)

```python
#Dimension del df
print(f"Dimensiones del DataFrame (Filas X Columnas): {df_titanic.shape} ")
```

```python
#Columnas
print(df_titanic.columns)
```

```python
df_titanic.info()
```

```python
display(df_titanic.isnull().sum())
```

```python
print("Limpieza básica de valores faltantes")

df_titanic["Age"] = df_titanic["Age"].fillna(df_titanic["Age"].mean())


print(f"Se rellena columna edad con promedio: {df_titanic["Age"].mean():.0f}")
```

```python
display(df_titanic.isnull().sum())
```

```python
most_frequent_embarked = df_titanic["Embarked"].mode()[0]
print(most_frequent_embarked)

df_titanic["Embarked"] = df_titanic["Embarked"].fillna(most_frequent_embarked)
print(f"Valores nulos en 'Embarked' rellenados con la moda ('{most_frequent_embarked}').")
```

```python
display(df_titanic.isnull().sum())
```

```python
df_titanic["Has_Cabin"] = df_titanic['Cabin'].notnull()

display(df_titanic["Has_Cabin"])
```

```python
#Borramos columna Cabin
df_titanic = df_titanic.drop('Cabin', axis = 1)
```

```python
display(df_titanic.isnull().sum())
```

```python
display(df_titanic.describe(include = 'object'))#Variables categorias Object


display(df_titanic.describe()) # Variables numéricas
```

```python
#Conteo de sobrevivientes

print(df_titanic['Survived'].value_counts())

print((df_titanic['Survived'].value_counts(normalize = True) * 100).round(2))
```

```python
print(df_titanic['Pclass'].value_counts())

print((df_titanic['Pclass'].value_counts(normalize = True) * 100).round(2))
```

```python
print(df_titanic['Sex'].value_counts())

print((df_titanic['Sex'].value_counts(normalize = True) * 100).round(2))
```

```python
print(f"Edad promedio: {df_titanic["Age"].mean():.2f}")
```

```python
df_sobrevivientes = df_titanic[df_titanic["Survived"] == 1]
display(df_sobrevivientes.head())

print("Numero sobrevivientes:", len(df_sobrevivientes))
print("Edad promedio sobrevivientes:", df_sobrevivientes["Age"].mean().round(2), "años")
```

```python
print("Tasa de supervivencia por Género (Análisis objetivo)")


tasa_hombres  = df_titanic[df_titanic["Sex"]  == 'male']['Survived'].mean() * 100
print(f"Tasa de supervivencia hombres: {tasa_hombres:.2f}%")

tasa_mujeres  = df_titanic[df_titanic["Sex"]  == 'female']['Survived'].mean() * 100
print(f"Tasa de supervivencia mujeres: {tasa_mujeres:.2f}%")
```

```python
print("Tasa de supervivencia por Clase de Boleto (Análisis objetivo)")


tasa_pclass1  = df_titanic[df_titanic["Pclass"]  == 1]['Survived'].mean() * 100
print(f"Tasa de supervivencia pclass1: {tasa_pclass1:.2f}%")

tasa_pclass2  = df_titanic[df_titanic["Pclass"]  == 2]['Survived'].mean() * 100
print(f"Tasa de supervivencia pclass2: {tasa_pclass2:.2f}%")

tasa_pclass3  = df_titanic[df_titanic["Pclass"]  == 3]['Survived'].mean() * 100
print(f"Tasa de supervivencia pclass3: {tasa_pclass3:.2f}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x = 'Sex', y = 'Survived', data = df_titanic)

plt.title("Tasa de supervivencia por Género")
plt.xlabel("Género")
plt.ylabel("Tasa de supervivencia")
#Mostrar el gráfica
plt.show()
```

```python
sns.barplot(x = 'Pclass', y = 'Survived', data = df_titanic)

plt.title("Tasa de supervivencia por Clase de Pasajero")
plt.xlabel("Clase de pasajero")
plt.ylabel("Tasa de supervivencia")
#Mostrar el gráfica
plt.show()
```
